```python
dataframe = [lebron, jordan, hakeem, chuck, david, karl]
```

```python
# 2012-13
d1213 = pd.concat([df.iloc[[0]] for df in dataframe]).reset_index(drop=True)

# 2009-10
d0910 = pd.concat([df.iloc[[1]] for df in dataframe]).reset_index(drop=True)

# 2008-09
d0809 = pd.concat([df.iloc[[2]] for df in dataframe]).reset_index(drop=True)

# 2011-12
d1112 = pd.concat([df.iloc[[3]] for df in dataframe]).reset_index(drop=True)

# 2010-11
d1011 = pd.concat([df.iloc[[4]] for df in dataframe]).reset_index(drop=True)

d = {'2008-09':d0809, '2009-10':d0910, '2010-11':d1011, '2011-12':d1112, '2012-13':d1213}
```

```python
d1213
```

```python
leagueavs = pd.read_csv('LeagueAveragesByYear.csv')
```

```python
seasons = ['2008-09', '2009-10', '2010-11', '2011-12', '2012-13']
```

```python
for season in seasons:
    yearavs = leagueavs[(leagueavs['Season'] == season)].reset_index(drop=True)
    possper48 = yearavs['Pace']
    dseason = d[season]
    mult_factor = possper48 / (48 * 100)
    for i in dseason.index:
        mf = dseason['2.MP'][i] * mult_factor
        dseason['2.FG'][i] = mf * dseason['4.FG'][i]
        dseason['2.FGA'][i] = mf * dseason['4.FGA'][i]
        dseason['2.3P'][i] = mf * dseason['4.3P'][i]
        dseason['2.3PA'][i] = mf * dseason['4.3PA'][i]
        dseason['2.2P'][i] = mf * dseason['4.2P'][i]
        dseason['2.2PA'][i] = mf * dseason['4.2PA'][i]
        dseason['2.FT'][i] = mf * dseason['4.FT'][i]
        dseason['2.FTA'][i] = mf * dseason['4.FTA'][i]
        dseason['2.ORB'][i] = mf * dseason['4.ORB'][i]
        dseason['2.DRB'][i] = mf * dseason['4.DRB'][i]
        dseason['2.TRB'][i] = mf * dseason['4.TRB'][i]
        dseason['2.AST'][i] = mf * dseason['4.AST'][i]
        dseason['2.STL'][i] = mf * dseason['4.STL'][i]
        dseason['2.BLK'][i] = mf * dseason['4.BLK'][i]
        dseason['2.TOV'][i] = mf * dseason['4.TOV'][i]
        dseason['2.PF'][i] = mf * dseason['4.PF'][i]
        dseason['2.PTS'][i] = mf * dseason['4.PTS'][i]

        dseason['1.FG'][i] = dseason['G'][i] * dseason['2.FG'][i]
        dseason['1.FGA'][i] = dseason['G'][i] * dseason['2.FGA'][i]
        dseason['1.3P'][i] = dseason['G'][i] * dseason['2.3P'][i]
        dseason['1.3PA'][i] = dseason['G'][i] * dseason['2.3PA'][i]
        dseason['1.2P'][i] = dseason['G'][i] * dseason['2.2P'][i]
        dseason['1.2PA'][i] = dseason['G'][i] * dseason['2.2PA'][i]
        dseason['1.FT'][i] = dseason['G'][i] * dseason['2.FT'][i]
        dseason['1.FTA'][i] = dseason['G'][i] * dseason['2.FTA'][i]
        dseason['1.ORB'][i] = dseason['G'][i] * dseason['2.ORB'][i]
        dseason['1.DRB'][i] = dseason['G'][i] * dseason['2.DRB'][i]
        dseason['1.TRB'][i] = dseason['G'][i] * dseason['2.TRB'][i]
        dseason['1.AST'][i] = dseason['G'][i] * dseason['2.AST'][i]
        dseason['1.STL'][i] = dseason['G'][i] * dseason['2.STL'][i]
        dseason['1.BLK'][i] = dseason['G'][i] * dseason['2.BLK'][i]
        dseason['1.TOV'][i] = dseason['G'][i] * dseason['2.TOV'][i]
        dseason['1.PF'][i] = dseason['G'][i] * dseason['2.PF'][i]
        dseason['1.PTS'][i] = dseason['G'][i] * dseason['2.PTS'][i]
```

```python
histo_test2 = pd.DataFrame()
histo_test2_1213 = pd.DataFrame()
for i in d.values():
    histo_test2 = pd.concat([histo_test2, i], ignore_index=True)
histo_test2_1213 = histo_test2.iloc[24:]
histo_test2 = histo_test2.iloc[:24]
```
