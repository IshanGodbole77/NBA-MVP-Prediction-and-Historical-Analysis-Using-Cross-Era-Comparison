```python
import pandas as pd
```

```python
dfe1 = pd.read_csv('ModernEval.csv')
dfe1
```

```python
dfe2 = pd.read_csv('MedievalEval.csv')
dfe2
```

```python
dfe3 = pd.read_csv('AncientEval.csv')
dfe3
```

```python
df = pd.merge(dfe1, dfe2, on = ['Year', 'Player', 'Age', 'MVP'])
```

```python
df = pd.merge(df, dfe3, on = ['Year', 'Player', 'Age', 'MVP'])
```

```python
df.rename(columns = {'Predicted MVP Share_x':'ModernEval', 'Predicted MVP Share_y':'MedievalEval', 'Predicted MVP Share':'AncientEval'})
```

```python
df['TotalEval'] = df.iloc[:, 4:7].mean(axis = 1)
```

```python
df
```

```python
df = df.sort_values('TotalEval', ascending=False).reset_index(drop=True)
```

```python
df
```

```python
df.to_csv('Results.csv', index=False)
```
